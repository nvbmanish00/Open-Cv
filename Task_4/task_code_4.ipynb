```python
import cv2
import mediapipe as mp

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Open video file
cap = cv2.VideoCapture('D:/DBDA/code_task_4/task_4_video.mp4')

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use appropriate codec
output_video = cv2.VideoWriter('output_task_4.mp4', fourcc, fps, (width, height))

# Initialize MediaPipe Pose model
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        # Read video frame
        success, image = cap.read()
        if not success:
            break

        # Process image with MediaPipe Pose model
        results = pose.process(image)

        # Draw pose landmarks and skeleton on image
        annotated_image = image.copy()
        mp_drawing.draw_landmarks(annotated_image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        # Write the annotated frame to the output video
        output_video.write(annotated_image)

        # Display image with pose landmarks and skeleton
        cv2.imshow('MediaPipe Pose Estimation', annotated_image)
        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

# Release video file, output video, and close windows
cap.release()
output_video.release()
cv2.destroyAllWindows()
```
